```python
def drop_useless_columns(df):
    dff = deepcopy(df[df.iloc[:, 1] == df.iloc[0, 1]])
    dff1 = dff[dff.columns[:4]]
    dff2 = dff[dff.columns[-1]]
    dff = pd.concat([dff1, dff2], axis=1)
    return dff.drop(1, axis=1)
```

```python
def split_by_spawned(df):
    list_of_games = []
    split_indexes = df[df[3] == 'Spawned'].index
    split_indexes = list(split_indexes)
    if len(split_indexes) != 0:
        list_of_games.append(df.loc[:split_indexes[0]])
    for i in range(len(split_indexes) - 2):
        list_of_games.append(df.loc[split_indexes[i]:split_indexes[i+1]])
    return (list_of_games, len(list_of_games))
```

```python
def multiply_target(targ_array, multiplying_list):
    ans = np.array([[0,0,0,0,0]], dtype=float)
    for i in range(len(multiplying_list)):
       ans = np.concatenate((ans, np.repeat(targ_array[i].reshape((1,-1)), multiplying_list[i], axis=0)), axis=0)
    return ans
```

#***Первый подход***# 

Относительно интерпретации действий роботов можно выдвинуть два альтернативных предположения

1) Роботы действуют сравнительно независимо от действий игрока, то есть действия роботов не сильно зависят от человека.

2) Роботы "подстраиваются под игрока" - их действия в большой степени зависят от психотеста игрока.

Начнем с первой - она диктует, что при предобработке надо оставить в логах только действия человека, поскольку действия роботов создают шум.

```python
def drop_useless_rows(df):
    l = list(df[2].unique())
    l.sort(key=lambda x: len(x)) # l -- list of unique players
    player = l[0]
    return df[df[2] == player]
```

```python
def g(df):
    l = list(df[2].unique())
    l.sort(key=lambda x: len(x))
    player = l[0]
    actions = df[df[2] == player][3].value_counts()
    return actions
```

```python
def h(df):
    my_amazing_df = pd.DataFrame()
    my_amazing_df['Winner'] = pd.Series(1 * (df.get('TakeOff', 0) > 0))
    meaningful_actions = set(df.index) - set(['TakeOff', 'exitTP', 'EnterTP', 'Escaped', 'Saved', 'Spawned', 'Deactivated', 'Reset']) # get rid of excess win-related actions
    for ma in meaningful_actions:
        my_amazing_df[ma] = pd.Series(df.get(ma, 0))
    return my_amazing_df
```

```python
X = pd.DataFrame()
multiplying_target_list = [] 
for filename in os.listdir('psihMLpsih'):
    if filename[:6] == 'social':
        print(filename)
        df_f = pd.read_csv('psihMLpsih/' + filename, header=None, names=np.arange(17))
        df_f = drop_useless_columns(df_f)
        df_f = drop_useless_rows(df_f)
        df_list, num_of_games = split_by_spawned(df_f)
        multiplying_target_list.append(num_of_games)
        for i in range(len(df_list)):
            X_addition = h(g(df_list[i]))
            X = pd.concat([X, X_addition], axis=0)
```

Вспомним функцию потерь - мы пользуемся коэффициентом детерминации, 

$R  = 1 - \frac{\sum\limits_{i=1}^{n}(\hat{y}_i - y_i)^2}{\sum\limits_{i=1}^{n}(\overline{y}-y_i)^2}$, где $\overline{y}$ - средний таргет и $\hat{y}_i$, $y_i$ - предсказанные и истинные таргеты соответственно.

Коэффициент детерминации имеет смысл объясняемой моделью доли дисперсии - действительно, в знаменателе стоит дисперсия оптимального константного предсказания, а в числителе - дисперсия предсказания модели.

Пока что полученные линейными моделями результаты хуже простого предсказания средним (в отличие от наивного гауссовского байеса, дающего неотрицательные и часто положительные качества).

#***Второй подход***#

Выше мы предполагали, что действия робов сильно не зависят от действий игрока, и потому рассмотрение действий роботов как признаков внесет в игровой процесс шум. Здесь будут учитываться действия роботов.

```python
def drop_useless_rows(df):
    l = list(df[2].unique())
    l.sort(key=lambda x: len(x)) # l -- list of unique players
    player = l[0]
    return df[(df[2] == player) | (df[3] != 'Spawned')]
```

```python
drop_useless_rows(df)
```

```python
def g(df):
    l = list(df[2].unique())
    l.sort(key=lambda x: len(x))
    player = l[0]
    actions_player = df[df[2] == player][3].value_counts()
    actions_nonplayer = df[df[2] != player][3].value_counts()
    new_index = []
    for idx in range(len(actions_nonplayer.index)):
      new_index.append('NP_' + actions_nonplayer.index[idx])
    actions_nonplayer.index = new_index
    return pd.concat((actions_player, actions_nonplayer))
```

```python
def h(df):
    my_amazing_df = pd.DataFrame()
    my_amazing_df['Winner'] = pd.Series(1 * (df.get('TakeOff', 0) > 0))
    meaningful_actions = set(df.index) - set(['TakeOff', 'exitTP', 'EnterTP', 'Escaped', 'Saved', 'Spawned', 'Deactivated', 
                                              'Reset']) - set(['NP_TakeOff', 'NP_exitTP', 'NP_EnterTP', 'NP_Escaped', 'NP_Saved', 'NP_Spawned',
                                              'NP_Deactivated', 'NP_Reset']) # get rid of excess win-related actions
    for ma in meaningful_actions:
        my_amazing_df[ma] = pd.Series(df.get(ma, 0))
    return my_amazing_df
```
